# 1. Word audio file code ğŸ” 

```python
# í•„ìš”í•œ ë¼ì´ë¸ŒëŸ¬ë¦¬ ì„¤ì¹˜
!pip install gtts gradio

# ë¼ì´ë¸ŒëŸ¬ë¦¬ import
from gtts import gTTS
import gradio as gr

# ë‹¨ì–´ ë¦¬ìŠ¤íŠ¸ ì •ì˜ (ë‹¨ì–´ì™€ ëœ»ì„ í•¨ê»˜ ì €ì¥)
words_with_meanings = [
    ("nestle", "í¬ê·¼íˆ ì•‰ë‹¤"),
    ("lush", "ë¬´ì„±í•œ"),
    ("explore", "íƒí—˜í•˜ë‹¤"),
    ("soil", "í† ì–‘"),
    ("tough", "íŠ¼íŠ¼í•œ"),
    ("dry", "ê±´ì¡°í•œ"),
    ("hardy", "íŠ¼íŠ¼í•œ"),
    ("peddler", "í–‰ìƒì¸"),
    ("wondrous", "ê²½ì´ë¡œìš´"),
    ("curious", "í˜¸ê¸°ì‹¬ì´ ë§ì€"),
    ("approach", "ì ‘ê·¼í•˜ë‹¤"),
    ("packet", "ê¾¸ëŸ¬ë¯¸"),
    ("skeptical", "íšŒì˜ì ì¸"),
    ("enthusiasm", "ì—´ì •"),
    ("sprouts", "ì‹¹"),
    ("vibrantly", "í™œê¸°ì°¨ê²Œ"),
    ("weed", "ì¡ì´ˆ"),
    ("transform", "ë³€í˜•í•˜ë‹¤"),
    ("marvel", "ê²½ì´ë¡œì›€"),
    ("prosper", "ë²ˆì˜í•˜ë‹¤"),
    ("stubborn", "ê³ ì§‘ ì„¼"),
    ("inspire", "ì˜ê°ì„ ì£¼ë‹¤"),
    ("unity", "ë‹¨ê²°"),
    ("beloved", "ì‚¬ë‘í•˜ëŠ”"),
    ("remind", "ìƒê¸°ì‹œí‚¤ë‹¤")
]

# ë°œìŒ íŒŒì¼ ìƒì„± í•¨ìˆ˜ ì •ì˜
def generate_audio(word):
    tts = gTTS(word, lang='en')       # ì…ë ¥ëœ ë‹¨ì–´ì— ëŒ€í•´ ì˜ì–´ë¡œ ë°œìŒ íŒŒì¼ ìƒì„±
    audio_file = f"{word}.mp3"        # ìƒì„±ëœ ì˜¤ë””ì˜¤ íŒŒì¼ ì´ë¦„
    tts.save(audio_file)              # ìƒì„±ëœ ë°œìŒ íŒŒì¼ ì €ì¥
    return audio_file                 # ìƒì„±ëœ ì˜¤ë””ì˜¤ íŒŒì¼ ê²½ë¡œ ë°˜í™˜

# Gradio ì¸í„°í˜ì´ìŠ¤ ì„¤ì •
def audio_generation_interface(word_with_meaning):
    word, meaning = word_with_meaning.split(" - ")
    audio_file = generate_audio(word) # ì…ë ¥ëœ ë‹¨ì–´ì— ëŒ€í•œ ë°œìŒ íŒŒì¼ ìƒì„±
    return audio_file, f"{word} - {meaning}" # ìƒì„±ëœ ë°œìŒ íŒŒì¼ ê²½ë¡œ ë° ë‹¨ì–´ì™€ ëœ» ë°˜í™˜

# ê° ë‹¨ì–´ì— ëŒ€í•œ ì„ íƒ ì˜µì…˜ ìƒì„±
options = [f"{word} - {meaning}" for word, meaning in words_with_meanings]

# Gradio ì¸í„°í˜ì´ìŠ¤ ìƒì„±
iface = gr.Interface(
    fn=audio_generation_interface,
    inputs=gr.Dropdown(choices=options, label="Select a word"),
    outputs=[gr.Audio(type="filepath", label="Pronunciation"), gr.Textbox(label="Word and Meaning")],
    title="ğŸ—£ï¸ Word Pronunciation Practice",
    description="Select a word from the dropdown menu to listen to its pronunciation and see its meaning."
)

# ì¸í„°í˜ì´ìŠ¤ ì‹¤í–‰
iface.launch(share=True)
```

```python
# 1. í•„ìš”í•œ íŒ¨í‚¤ì§€ ì„¤ì¹˜
!pip install gradio matplotlib pillow

# 2. ëŒë¦¼íŒ ì½”ë“œ ì‘ì„±
import random
import matplotlib.pyplot as plt
import numpy as np
import gradio as gr
from PIL import Image
from matplotlib.patches import Wedge

# ëŒë¦¼íŒì— í‘œì‹œí•  ë‹¨ì–´ë“¤
words = ["love", "death", "storm", "robot"]
colors = ["#ff9999","#66b3ff","#99ff99","#ffcc99"]

# ëŒë¦¼íŒ ê·¸ë¦¬ê¸° í•¨ìˆ˜
def draw_wheel(angle):
    fig, ax = plt.subplots(figsize=(6, 6), subplot_kw={'aspect': 'equal'})
    wedges = []

    for i, word in enumerate(words):
        wedges.append(Wedge(center=(0, 0), r=1, theta1=i * 90 + angle, theta2=(i + 1) * 90 + angle, facecolor=colors[i]))

    for wedge in wedges:
        ax.add_patch(wedge)

    ax.set_xlim(-1, 1)
    ax.set_ylim(-1, 1)
    ax.axis('off')

    plt.close(fig)
    return fig

# ëŒë¦¼íŒ ì• ë‹ˆë©”ì´ì…˜ ìƒì„± í•¨ìˆ˜
def spin_wheel():
    frames = []
    total_frames = 30
    random_angle = random.randint(0, 359)

    for i in range(total_frames):
        angle = (i * (random_angle / total_frames)) % 360
        fig = draw_wheel(angle)
        buf = fig.canvas.tostring_rgb()
        ncols, nrows = fig.canvas.get_width_height()
        image = np.frombuffer(buf, dtype=np.uint8).reshape(nrows, ncols, 3)
        frames.append(Image.fromarray(image))

    # ë§ˆì§€ë§‰ í”„ë ˆì„ì— ìµœì¢… ìœ„ì¹˜ì—ì„œ ì„ íƒëœ ë‹¨ì–´ í‘œì‹œ
    fig = draw_wheel(random_angle)
    buf = fig.canvas.tostring_rgb()
    ncols, nrows = fig.canvas.get_width_height()
    image = np.frombuffer(buf, dtype=np.uint8).reshape(nrows, ncols, 3)
    final_image = Image.fromarray(image)
    frames.append(final_image)

    # GIFë¡œ ì €ì¥
    gif_path = "/mnt/data/spinning_wheel.gif"
    frames[0].save(gif_path, save_all=True, append_images=frames[1:], optimize=False, duration=100, loop=0)

    return gif_path

# Gradio ì¸í„°í˜ì´ìŠ¤ ì„¤ì •
iface = gr.Interface(
    fn=spin_wheel,
    inputs=None,
    outputs="image",
    title="Word Spinner",
    description="Spin the wheel to get one of the words: love, death, storm, robot."
)

# Gradio ì¸í„°í˜ì´ìŠ¤ ì‹¤í–‰
iface.launch()
```
