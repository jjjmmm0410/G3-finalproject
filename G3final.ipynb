# 1. Word audio file code ğŸ” 

```python
# í•„ìš”í•œ ë¼ì´ë¸ŒëŸ¬ë¦¬ ì„¤ì¹˜
!pip install gtts gradio

# ë¼ì´ë¸ŒëŸ¬ë¦¬ import
from gtts import gTTS
import gradio as gr

# ë‹¨ì–´ ë¦¬ìŠ¤íŠ¸ ì •ì˜ (ë‹¨ì–´ì™€ ëœ»ì„ í•¨ê»˜ ì €ì¥)
words_with_meanings = [
    ("nestle", "í¬ê·¼íˆ ì•‰ë‹¤"),
    ("lush", "ë¬´ì„±í•œ"),
    ("explore", "íƒí—˜í•˜ë‹¤"),
    ("soil", "í† ì–‘"),
    ("tough", "íŠ¼íŠ¼í•œ"),
    ("dry", "ê±´ì¡°í•œ"),
    ("hardy", "íŠ¼íŠ¼í•œ"),
    ("peddler", "í–‰ìƒì¸"),
    ("wondrous", "ê²½ì´ë¡œìš´"),
    ("curious", "í˜¸ê¸°ì‹¬ì´ ë§ì€"),
    ("approach", "ì ‘ê·¼í•˜ë‹¤"),
    ("packet", "ê¾¸ëŸ¬ë¯¸"),
    ("skeptical", "íšŒì˜ì ì¸"),
    ("enthusiasm", "ì—´ì •"),
    ("sprouts", "ì‹¹"),
    ("vibrantly", "í™œê¸°ì°¨ê²Œ"),
    ("weed", "ì¡ì´ˆ"),
    ("transform", "ë³€í˜•í•˜ë‹¤"),
    ("marvel", "ê²½ì´ë¡œì›€"),
    ("prosper", "ë²ˆì˜í•˜ë‹¤"),
    ("stubborn", "ê³ ì§‘ ì„¼"),
    ("inspire", "ì˜ê°ì„ ì£¼ë‹¤"),
    ("unity", "ë‹¨ê²°"),
    ("beloved", "ì‚¬ë‘í•˜ëŠ”"),
    ("remind", "ìƒê¸°ì‹œí‚¤ë‹¤")
]

# ë°œìŒ íŒŒì¼ ìƒì„± í•¨ìˆ˜ ì •ì˜
def generate_audio(word):
    tts = gTTS(word, lang='en')       # ì…ë ¥ëœ ë‹¨ì–´ì— ëŒ€í•´ ì˜ì–´ë¡œ ë°œìŒ íŒŒì¼ ìƒì„±
    audio_file = f"{word}.mp3"        # ìƒì„±ëœ ì˜¤ë””ì˜¤ íŒŒì¼ ì´ë¦„
    tts.save(audio_file)              # ìƒì„±ëœ ë°œìŒ íŒŒì¼ ì €ì¥
    return audio_file                 # ìƒì„±ëœ ì˜¤ë””ì˜¤ íŒŒì¼ ê²½ë¡œ ë°˜í™˜

# Gradio ì¸í„°í˜ì´ìŠ¤ ì„¤ì •
def audio_generation_interface(word_with_meaning):
    word, meaning = word_with_meaning.split(" - ")
    audio_file = generate_audio(word) # ì…ë ¥ëœ ë‹¨ì–´ì— ëŒ€í•œ ë°œìŒ íŒŒì¼ ìƒì„±
    return audio_file, f"{word} - {meaning}" # ìƒì„±ëœ ë°œìŒ íŒŒì¼ ê²½ë¡œ ë° ë‹¨ì–´ì™€ ëœ» ë°˜í™˜

# ê° ë‹¨ì–´ì— ëŒ€í•œ ì„ íƒ ì˜µì…˜ ìƒì„±
options = [f"{word} - {meaning}" for word, meaning in words_with_meanings]

# Gradio ì¸í„°í˜ì´ìŠ¤ ìƒì„±
iface = gr.Interface(
    fn=audio_generation_interface,
    inputs=gr.Dropdown(choices=options, label="Select a word"),
    outputs=[gr.Audio(type="filepath", label="Pronunciation"), gr.Textbox(label="Word and Meaning")],
    title="ğŸ—£ï¸ Word Pronunciation Practice",
    description="Select a word from the dropdown menu to listen to its pronunciation and see its meaning."
)

# ì¸í„°í˜ì´ìŠ¤ ì‹¤í–‰
iface.launch(share=True)
```

```python

import random
import gradio as gr
from PIL import Image

# ì„ì¸ ì´ë¯¸ì§€ë¥¼ ì €ì¥í•  ë¦¬ìŠ¤íŠ¸
shuffled_images = []
original_images = []

# ì´ë¯¸ì§€ ì—…ë¡œë“œ ë° ì„ê¸° í•¨ìˆ˜
def upload_and_shuffle(image1, image2, image3, image4):
    global shuffled_images, original_images
    images = [image1, image2, image3, image4]
    original_images = images.copy()
    random.shuffle(images)
    shuffled_images = images
    return images

# ìˆœì„œ í™•ì¸ í•¨ìˆ˜
def check_order(slot1, slot2, slot3, slot4):
    user_order = [slot1 - 1, slot2 - 1, slot3 - 1, slot4 - 1]  # Adjust for 1-based index
    correct_order = [shuffled_images.index(img) for img in original_images]  # Correct order based on original images
    correct = user_order == correct_order
    return (
        "ì •ë‹µì…ë‹ˆë‹¤!" if correct else "í‹€ë ¸ìŠµë‹ˆë‹¤. ì˜¬ë°”ë¥¸ ìˆœì„œëŠ”:",
        original_images[0], original_images[1],
        original_images[2], original_images[3]
    )

# Gradio ì¸í„°í˜ì´ìŠ¤ ìƒì„±
def create_interface():
    with gr.Blocks() as demo:
        with gr.Row():
            image1 = gr.Image(label="ì´ë¯¸ì§€ 1 ì—…ë¡œë“œ", type="pil")
            image2 = gr.Image(label="ì´ë¯¸ì§€ 2 ì—…ë¡œë“œ", type="pil")
            image3 = gr.Image(label="ì´ë¯¸ì§€ 3 ì—…ë¡œë“œ", type="pil")
            image4 = gr.Image(label="ì´ë¯¸ì§€ 4 ì—…ë¡œë“œ", type="pil")

        upload_button = gr.Button("Upload and Shuffle Images")
        output_image1 = gr.Image(label="ì„ì¸ ì´ë¯¸ì§€ 1")
        output_image2 = gr.Image(label="ì„ì¸ ì´ë¯¸ì§€ 2")
        output_image3 = gr.Image(label="ì„ì¸ ì´ë¯¸ì§€ 3")
        output_image4 = gr.Image(label="ì„ì¸ ì´ë¯¸ì§€ 4")

        with gr.Row():
            slot1 = gr.Dropdown(choices=[1, 2, 3, 4], label="Slot 1")
            slot2 = gr.Dropdown(choices=[1, 2, 3, 4], label="Slot 2")
            slot3 = gr.Dropdown(choices=[1, 2, 3, 4], label="Slot 3")
            slot4 = gr.Dropdown(choices=[1, 2, 3, 4], label="Slot 4")

        submit_button = gr.Button("Submit")
        result = gr.Label()
        correct_order_image1 = gr.Image(label="ì •ë‹µ ìˆœì„œ ì´ë¯¸ì§€ 1")
        correct_order_image2 = gr.Image(label="ì •ë‹µ ìˆœì„œ ì´ë¯¸ì§€ 2")
        correct_order_image3 = gr.Image(label="ì •ë‹µ ìˆœì„œ ì´ë¯¸ì§€ 3")
        correct_order_image4 = gr.Image(label="ì •ë‹µ ìˆœì„œ ì´ë¯¸ì§€ 4")

        upload_button.click(upload_and_shuffle, inputs=[image1, image2, image3, image4], outputs=[output_image1, output_image2, output_image3, output_image4])
        submit_button.click(check_order, inputs=[slot1, slot2, slot3, slot4], outputs=[result, correct_order_image1, correct_order_image2, correct_order_image3, correct_order_image4])

    return demo

# Gradio ì¸í„°í˜ì´ìŠ¤ ì‹¤í–‰
demo = create_interface()
demo.launch()
```

```python
# Install necessary packages
!pip install gradio

import random
import gradio as gr

def ladder_game(participants, outcomes):
    if len(participants) != len(outcomes):
        return "The number of participants must be equal to the number of outcomes."

    random.shuffle(outcomes)
    results = dict(zip(participants, outcomes))

    return results

def gradio_ladder_game():
    participants = ["Team 1", "Team 2", "Team 3", "Team 4"]
    outcomes = ["love", "robot", "storm", "death"]

    results = ladder_game(participants, outcomes)
    return results

iface = gr.Interface(
    fn=gradio_ladder_game,
    inputs=None,
    outputs="json",
    title="Drawing lots!ğŸŒ±",
    description="Each team (Team 1, Team 2, Team 3, Team 4) will get one of the outcomes: love, robot, storm, death."
)

iface.launch()
```
