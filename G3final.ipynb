```python

# í•„ìš”í•œ ë¼ì´ë¸ŒëŸ¬ë¦¬ ì„¤ì¹˜
!pip install gtts gradio

# ë¼ì´ë¸ŒëŸ¬ë¦¬ import
from gtts import gTTS
import gradio as gr

# ë‹¨ì–´ ë¦¬ìŠ¤íŠ¸ ì •ì˜
words = [
    "nestle", "lush", "explore", "soil", "tough", "dry", "hardy", "peddler",
    "wondrous", "curious", "approach", "packet", "skeptical", "enthusiasm",
    "sprouts", "vibrantly", "weed", "transform", "marvel", "prosper", "stubborn",
    "inspire", "unity", "beloved", "remind"
]

# ë°œìŒ íŒŒì¼ ìƒì„± í•¨ìˆ˜ ì •ì˜
def generate_audio(word):
    tts = gTTS(word, lang='en')       # ì…ë ¥ëœ ë‹¨ì–´ì— ëŒ€í•´ ì˜ì–´ë¡œ ë°œìŒ íŒŒì¼ ìƒì„±
    audio_file = f"{word}.mp3"        # ìƒì„±ëœ ì˜¤ë””ì˜¤ íŒŒì¼ ì´ë¦„
    tts.save(audio_file)              # ìƒì„±ëœ ë°œìŒ íŒŒì¼ ì €ì¥
    return audio_file                 # ìƒì„±ëœ ì˜¤ë””ì˜¤ íŒŒì¼ ê²½ë¡œ ë°˜í™˜

# Gradio ì¸í„°í˜ì´ìŠ¤ ì„¤ì •
def audio_generation_interface(word):
    audio_file = generate_audio(word) # ì…ë ¥ëœ ë‹¨ì–´ì— ëŒ€í•œ ë°œìŒ íŒŒì¼ ìƒì„±
    return audio_file                 # ìƒì„±ëœ ë°œìŒ íŒŒì¼ ê²½ë¡œ ë°˜í™˜

# Gradio ì¸í„°í˜ì´ìŠ¤ ìƒì„±
iface = gr.Interface(
    fn=audio_generation_interface,
    inputs=gr.Dropdown(choices=words, label="Select a word"),
    outputs=gr.Audio(type="filepath", label="Pronunciation"), # typeì„ 'filepath'ë¡œ ì„¤ì •
    title="Word Pronunciation Practice",
    description="Select a word from the dropdown menu to listen to its pronunciation."
)

# ì¸í„°í˜ì´ìŠ¤ ì‹¤í–‰
iface.launch(share=True)
```
